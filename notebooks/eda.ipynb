```python
hpg_visit_data = hpg_reserve.drop(columns = ['reserve_datetime']).groupby(by = ["hpg_store_id", "visit_datetime"]).sum().reset_index()
hpg_visit_data = hpg_visit_data.rename(columns={"visit_datetime": "visit_date", "reserve_visitors": "visitors"})
hpg_visit_data
```

```python
hpg_visit_data['visit_date'] = pd.to_datetime(hpg_visit_data['visit_date'])
hpg_visit_data.info()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['visit_date']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'visit_date'}, axis=1)
              .sort_values('visit_date', ascending=True))
  xs = counted['visit_date']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = hpg_visit_data.sort_values('visit_date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('visit_date')
_ = plt.ylabel('count()')
```

```python
visit_data = visit_data.rename(columns={"visit_date": "date"})
date_info = date_info.rename(columns={"calendar_date": "date"})
```

```python
visit_data['date'] = visit_data['date'].astype("string")
date_info['date'] = date_info['date'].astype("string")
```

```python
data = pd.merge(visit_data, date_info, on="date")
data = pd.merge(data, store_info, on="store_id")
data['date'] = pd.to_datetime(data['date'])
data
```

```python
from matplotlib import pyplot as plt
data['holiday_flg'].plot(kind='hist', bins=20, title='holiday_flg')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
store_id_relation = pd.read_csv("/root/restaurants/data/raw/store_id_relation.csv")
store_id_relation
```

```python
air = air_store_info.loc[air_store_info['store_id'] == store_id_relation.iloc[1].iloc[0]].reset_index(drop = True)
hpg = hpg_store_info.loc[hpg_store_info['store_id'] == store_id_relation.iloc[1].iloc[1]].reset_index(drop = True)
df = pd.concat([air, hpg], axis=1)

for i in range(150):
    air = air_store_info.loc[air_store_info['store_id'] == store_id_relation.iloc[i].iloc[0]].reset_index(drop = True)
    hpg = hpg_store_info.loc[hpg_store_info['store_id'] == store_id_relation.iloc[i].iloc[1]].reset_index(drop = True)
    df = pd.concat([df, pd.concat([air, hpg], axis=1)], ignore_index = True)

df.drop([0])
df
```
