# Reverse Integer

Division-based solution

```python
def division(value):
    new_value = 0

    while (value != 0):
        new_value = new_value * 10 + value % 10
        value = value // 10

    return new_value
```

String-based solution

```python
def string(value):
    value_str = str(value)

    new_value_str = value_str[::-1]

    return int(value_str[::-1])
```

```python
import random

small = [random.randint(0, 10**2) for i in range(100000)]
medium = [random.randint(10**2, 10**4) for i in range(100000)]
long_ = [random.randint(10**4, 10**8) for i in range(100000)]
long_long = [random.randint(10**8, 10**16) for i in range(100000)]
extra = [random.randint(10**16, 10**32) for i in range(100000)]
all = small + medium + long_ + long_long + extra

values = ['small', 'medium', 'long_', 'long_long', 'extra', 'all']
```

```python
import timeit

def check_time(reverse, values):
    times = list()

    for value in values:
      testcode = '''for i in ''' + value + ": " + reverse + '''(i)'''
      time = timeit.timeit(stmt=testcode, globals=globals(), number=10)
      times.append(time)
      print(value + ":", time)

    times.pop()
    return times
```

```python
str_time =check_time("string", values)
```

```python
div_time = check_time("division", values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10**2, 10**4, 10**8, 10**16, 10**32])

y1 = str_time
y2 = div_time

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='String-based solution', color='blue', marker='o')
plt.plot(x, y2, label='Division-based solution', color='red', marker='o')

plt.legend()

plt.xscale('log') 

plt.xlabel('x (log scale)')
plt.ylabel('y')

plt.grid(True, which="both", ls="--")

plt.show()
```

*Comparison:*
Both algorithms have the same time complexity of *O(log₁₀(n))*, where n is the number that needs to be reversed. But, based on the results of the *check_time()* function, which measures the execution time of a function, it can be observed that when the length of the numbers doubles, the execution time of the division-based algorithm increases accordingly and the string-based algorithm does not depend on the length of the numbers and is depends only on their quantity. String-based algorithm is shorter and easier to understand and implement.

# Custom object as a key in a dictionary

```python
class MyType:
    def __init__(self, value):
        self.value = value

    def __hash__(self):
        return hash(self.value)

    def __eq__(self, other):
        return self.value == other.value
```

```python
d = dict()
d[MyType(1)] = 458
d[MyType(5)] = 32
d[MyType(12)] = 6546
d[MyType(66)] = 78
d[MyType(8)] = 4
```
